# 2.1.7.2 cleaning the column name

1. at first the name of the 'yearOfRegistration','monthOfRegistration','notRepairedDamage','dateCreated' is converted directy by rewriting and assigning the column name
2. to convert the camelcase to snakecase a function name 'cameltosnake' is created
3. using for loop we converted the camelcase into snakecase into the col_list

```python
col_name=(autos.columns)
col_name
```

```python
col_name=['dateCrawled', 'name', 'seller', 'offerType', 'price', 'abtest',
       'vehicleType', 'registration_year', 'gearbox', 'powerPS', 'model',
       'odometer', 'registration_month', 'fuelType', 'brand',
       'unrepaired_damage', 'ad_created', 'nrOfPictures', 'postalCode',
       'lastSeen']

def cameltosnake (string):
    output= [string[0].lower()]
    for char in string[1:]:
        if ord(char) in range(65,90):
            output.append('_')
            output.append(char.lower())
        else :
            output.append(char)
    return str.join('',output)

for ele in col_name:
    col_name[col_name.index(ele)]= cameltosnake(ele)

autos.columns= col_name
autos.head()
```

# 2.1.7.3 initial exploration and cleaning

```python
autos.describe(include='all')
```

1. To Remove  ','   and   'km'   character from 'odometer' column
2. To change 'odometer' datatype from sting to int
3. To change column name 'odometer' to 'odometer_km'

```python
autos['odometer']= autos['odometer'].str.replace(',','')
autos['odometer']= autos['odometer'].str.replace('km','')
autos['odometer']= autos['odometer'].astype(int)
autos.rename(columns={'odometer':'odometer_km'},inplace=True)
autos
```

1. To Remove ',' and 'km' character from 'price' column
2. To change 'price' column datatype from string to int

```python
autos['price']= autos['price'].str.replace('$','')
autos['price']= autos['price'].str.replace(',','')
autos['price']= autos['price'].astype(int)
autos
```
